### Subtask 1: Correct years for given data set

```python
# splitting Date column in 2 columns---- one will have year part and other one will have date and month part
new_date_column=data["Date"].str.split("-", n = 1, expand = True)

Current_year='2023'

while (Current_year!='2034'):

    if Current_year=='2023':
        new_date_column[0].mask(new_date_column[0].iloc[331:692]== '2022' ,'2023', inplace=True)

    if Current_year=='2024':
        new_date_column[0].mask(new_date_column[0].iloc[692:942]== '2022' ,'2024', inplace=True)
    
    if Current_year=='2025':
        new_date_column[0].mask(new_date_column[0].iloc[942:1279]== '2022' ,'2025', inplace=True)

    if Current_year=='2026':
        new_date_column[0].mask(new_date_column[0].iloc[1279:1599]== '2022' ,'2026', inplace=True)
    
    if Current_year=='2027':
        new_date_column[0].mask(new_date_column[0].iloc[1599:1956]== '2022' ,'2027', inplace=True)
    
    if Current_year=='2028':
        new_date_column[0].mask(new_date_column[0].iloc[1956:2307]== '2022' ,'2028', inplace=True)
    
    if Current_year=='2029':
        new_date_column[0].mask(new_date_column[0].iloc[2307:2673]== '2022' ,'2029', inplace=True)
    
    if Current_year=='2030':
        new_date_column[0].mask(new_date_column[0].iloc[2673:3009]== '2022' ,'2030', inplace=True)
    
    if Current_year=='2031':
        new_date_column[0].mask(new_date_column[0].iloc[3009:3333]== '2022' ,'2031', inplace=True)
    
    if Current_year=='2032':
        new_date_column[0].mask(new_date_column[0].iloc[3333:3692]== '2022' ,'2032', inplace=True)
    
    if Current_year=='2033':
        new_date_column[0].mask(new_date_column[0].iloc[3692:3903]== '2022' ,'2033', inplace=True)    
    
    

    Current_year=str(int(Current_year)+1)  
```

```python
data['Date']=new_date_column[0] + '-' + new_date_column[1]
```
