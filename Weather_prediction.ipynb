### Removing Colons in every Column

```python
for col in data:
    data[col] = data[col].str.replace("'","", regex=True)
```

```python
data
```

### Subtask 1: Correct years for given data set

```python
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d', errors='coerce')
data['Date'].dtype

target_date = '2022-01-01'                         
target_date = pd.to_datetime(target_date)
index = data.index[data['Date'] == target_date].tolist()
year = 2021
max_year = 2033

for i in range(index[0], data.shape[0]):
    if data.at[i, 'Date'] == target_date:
        year += 1
        if year > max_year:
            break
    data.at[i, 'Date'] = data.at[i, 'Date'].replace(year=year)
```

### Subtask 2: Removal of duplicate rows and duplicate Columns


```python
duplicate = data[data.duplicated()]
data.duplicated()
```

```python
duplicate
```

```python
# deleting duplicate rows .Here we only have 1 duplicate row
data.drop_duplicates(keep='first',inplace=True)
```

```python
# dropping the irrelevent columns
data.drop(['Date1',' Month'], axis=1,inplace=True)
```
